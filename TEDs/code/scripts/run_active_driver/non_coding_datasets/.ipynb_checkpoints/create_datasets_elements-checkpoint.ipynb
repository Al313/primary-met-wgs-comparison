```python
import pandas as pd
import glob
import os
```

### Create dataset ActiveDriver

```python
l=[]
for filein in glob.glob("hg19_*.gz"):
    df = pd.read_csv(filein,sep="\t")
    name = os.path.basename(filein).split(".")[0].replace("hg19_","")
    df["CHROMOSOME"] = df["CHROMOSOME"].astype(str)
    df["chr"] = "chr"+df["CHROMOSOME"]
    df["id"] = df["GENE_ID"] + "_" + df["SYMBOL"]
    df[["chr","START","END","id"]].rename(columns={"START":"start","END":"end"}).to_csv(name+".tsv.gz",sep="\t",index=False,compression="gzip")
    l.append(df)
pd.concat(l)[["chr","START","END","id"]].rename(columns={"START":"start","END":"end"}).to_csv("nc_elements.tsv.gz",sep="\t",index=False,compression="gzip")
    
```
